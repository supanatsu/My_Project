#Data Preparation

```python
# à¹à¸›à¸¥à¸‡ Age à¹€à¸›à¹‡à¸™à¸•à¸±à¸§à¹€à¸¥à¸‚à¹à¸¥à¸°à¸à¸£à¸­à¸‡ outliers
dataset['Age'] = pd.to_numeric(dataset['Age'], errors='coerce')
dataset_cleaned = dataset[(dataset['Age'].notna()) &
                         (dataset['Age'] >= 0) &
                         (dataset['Age'] <= 120)].copy()

print(f"à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¹€à¸”à¸´à¸¡: {dataset.shape[0]} à¹à¸–à¸§")
print(f"à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¸«à¸¥à¸±à¸‡à¸—à¸³à¸„à¸§à¸²à¸¡à¸ªà¸°à¸­à¸²à¸”: {dataset_cleaned.shape[0]} à¹à¸–à¸§")
print(f"à¸¥à¸šà¸­à¸­à¸: {dataset.shape[0] - dataset_cleaned.shape[0]} à¹à¸–à¸§")
```

```python
# à¹€à¸•à¸´à¸¡à¸„à¹ˆà¸² missing à¹ƒà¸™ Age à¸”à¹‰à¸§à¸¢ median
median_age = dataset_cleaned['Age'].median()
dataset_cleaned['Age'].fillna(median_age, inplace=True)
print(f"\nà¹€à¸•à¸´à¸¡à¸„à¹ˆà¸² Age à¸—à¸µà¹ˆà¸«à¸²à¸¢à¸”à¹‰à¸§à¸¢ median = {median_age}")

print("\nà¸à¸²à¸£à¸à¸£à¸°à¸ˆà¸²à¸¢à¸‚à¸­à¸‡ Age à¸«à¸¥à¸±à¸‡à¸—à¸³à¸„à¸§à¸²à¸¡à¸ªà¸°à¸­à¸²à¸”:")
display(dataset_cleaned['Age'].describe())
```

#Preparation à¸‚à¸­à¸‡ favorite

```python
# à¸à¸³à¸«à¸™à¸”à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ features à¹à¸¥à¸° target
features_cols_fav = ['Age', 'gender', 'occupation', 'education', 'activity']
target_col_fav = 'favorite'

# à¹à¸¢à¸à¸‚à¹‰à¸­à¸¡à¸¹à¸¥: à¸Šà¸¸à¸”à¸—à¸µà¹ˆà¹„à¸¡à¹ˆà¸¡à¸µ 'other' à¸ªà¸³à¸«à¸£à¸±à¸š train à¹à¸¥à¸°à¸Šà¸¸à¸”à¸—à¸µà¹ˆà¸¡à¸µ 'other' à¸ªà¸³à¸«à¸£à¸±à¸š impute
df_train_fav = dataset_cleaned[dataset_cleaned[target_col_fav] != 'other'].copy()
df_impute_fav = dataset_cleaned[dataset_cleaned[target_col_fav] == 'other'].copy()

print(f"\nà¸ˆà¸³à¸™à¸§à¸™à¸‚à¹‰à¸­à¸¡à¸¹à¸¥:")
print(f"  - Training set (favorite â‰  'other'): {len(df_train_fav)} à¹à¸–à¸§")
print(f"  - Impute set (favorite = 'other'): {len(df_impute_fav)} à¹à¸–à¸§")
```

```python
# à¹€à¸•à¸£à¸µà¸¢à¸¡ X à¹à¸¥à¸° y à¸ªà¸³à¸«à¸£à¸±à¸š favorite
X_fav = df_train_fav[features_cols_fav].copy()
y_fav = df_train_fav[target_col_fav].copy()

print(f"\nà¸£à¸¹à¸›à¸£à¹ˆà¸²à¸‡à¸‚à¸­à¸‡ X: {X_fav.shape}")
print(f"à¸£à¸¹à¸›à¸£à¹ˆà¸²à¸‡à¸‚à¸­à¸‡ y: {y_fav.shape}")
```

```python
# Label Encoding à¸ªà¸³à¸«à¸£à¸±à¸š target (y)
le_fav = LabelEncoder()
y_fav_encoded = le_fav.fit_transform(y_fav)

print(f"\nà¸„à¸¥à¸²à¸ªà¸‚à¸­à¸‡ favorite: {le_fav.classes_}")
print(f"à¸ˆà¸³à¸™à¸§à¸™à¸„à¸¥à¸²à¸ª: {len(le_fav.classes_)}")
```

```python
# à¹à¸šà¹ˆà¸‡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¹€à¸›à¹‡à¸™ train/test (80/20) à¸à¹ˆà¸­à¸™ encoding X
X_train_fav, X_test_fav, y_train_fav, y_test_fav = train_test_split(
    X_fav, y_fav_encoded, test_size=0.2, random_state=42, stratify=y_fav_encoded
)

print(f"\nà¸à¸²à¸£à¹à¸šà¹ˆà¸‡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥:")
print(f"  - Training: {len(X_train_fav)} à¹à¸–à¸§")
print(f"  - Testing: {len(X_test_fav)} à¹à¸–à¸§")
```

```python
# *** à¸ªà¸³à¸„à¸±à¸: One-Hot Encoding à¸«à¸¥à¸±à¸‡à¸ˆà¸²à¸ split à¹à¸¥à¹‰à¸§ ***
categorical_features_fav = ['gender', 'occupation', 'education', 'activity']

X_train_fav_encoded = pd.get_dummies(X_train_fav, columns=categorical_features_fav, drop_first=True)
X_test_fav_encoded = pd.get_dummies(X_test_fav, columns=categorical_features_fav, drop_first=True)
```

```python
# à¸ˆà¸±à¸”à¸à¸²à¸£à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œà¸—à¸µà¹ˆà¹„à¸¡à¹ˆà¸•à¸£à¸‡à¸à¸±à¸™
for col in X_train_fav_encoded.columns:
    if col not in X_test_fav_encoded.columns:
        X_test_fav_encoded[col] = 0

X_test_fav_encoded = X_test_fav_encoded[X_train_fav_encoded.columns]

print(f"\nà¸£à¸¹à¸›à¸£à¹ˆà¸²à¸‡à¸«à¸¥à¸±à¸‡ One-Hot Encoding:")
print(f"  - X_train: {X_train_fav_encoded.shape}")
print(f"  - X_test: {X_test_fav_encoded.shape}")
```

#Preparation à¸‚à¸­à¸‡ life style

```python
# à¸à¸³à¸«à¸™à¸”à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ features à¹à¸¥à¸° target
features_cols_life = ['Age', 'gender', 'occupation', 'education', 'activity']
target_col_life = 'lifestyle'

# à¹à¸¢à¸à¸‚à¹‰à¸­à¸¡à¸¹à¸¥
df_train_life = dataset_cleaned[dataset_cleaned[target_col_life] != 'other'].copy()
df_impute_life = dataset_cleaned[dataset_cleaned[target_col_life] == 'other'].copy()

print(f"\nà¸ˆà¸³à¸™à¸§à¸™à¸‚à¹‰à¸­à¸¡à¸¹à¸¥:")
print(f"  - Training set (lifestyle â‰  'other'): {len(df_train_life)} à¹à¸–à¸§")
print(f"  - Impute set (lifestyle = 'other'): {len(df_impute_life)} à¹à¸–à¸§")
```

```python
# à¹€à¸•à¸£à¸µà¸¢à¸¡ X à¹à¸¥à¸° y à¸ªà¸³à¸«à¸£à¸±à¸š lifestyle
X_life = df_train_life[features_cols_life].copy()
y_life = df_train_life[target_col_life].copy()

# Label Encoding à¸ªà¸³à¸«à¸£à¸±à¸š target
le_life = LabelEncoder()
y_life_encoded = le_life.fit_transform(y_life)

print(f"\nà¸„à¸¥à¸²à¸ªà¸‚à¸­à¸‡ lifestyle: {le_life.classes_}")
print(f"à¸ˆà¸³à¸™à¸§à¸™à¸„à¸¥à¸²à¸ª: {len(le_life.classes_)}")
```

```python
# à¹à¸šà¹ˆà¸‡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¹€à¸›à¹‡à¸™ train/test
X_train_life, X_test_life, y_train_life, y_test_life = train_test_split(
    X_life, y_life_encoded, test_size=0.2, random_state=42, stratify=y_life_encoded)

print(f"\nà¸à¸²à¸£à¹à¸šà¹ˆà¸‡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥:")
print(f"  - Training: {len(X_train_life)} à¹à¸–à¸§")
print(f"  - Testing: {len(X_test_life)} à¹à¸–à¸§")
```

```python
# One-Hot Encoding
categorical_features_life = ['gender', 'occupation', 'education', 'activity']

X_train_life_encoded = pd.get_dummies(X_train_life, columns=categorical_features_life, drop_first=True)
X_test_life_encoded = pd.get_dummies(X_test_life, columns=categorical_features_life, drop_first=True)
```

```python
# à¸ˆà¸±à¸”à¸à¸²à¸£à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œà¸—à¸µà¹ˆà¹„à¸¡à¹ˆà¸•à¸£à¸‡à¸à¸±à¸™
for col in X_train_life_encoded.columns:
    if col not in X_test_life_encoded.columns:
        X_test_life_encoded[col] = 0

X_test_life_encoded = X_test_life_encoded[X_train_life_encoded.columns]

print(f"\nà¸£à¸¹à¸›à¸£à¹ˆà¸²à¸‡à¸«à¸¥à¸±à¸‡ One-Hot Encoding:")
print(f"  - X_train: {X_train_life_encoded.shape}")
print(f"  - X_test: {X_test_life_encoded.shape}")
```
